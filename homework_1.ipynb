Как принять решение удалить или заполнить, есть несколько способов, все равно это искажение данных. Оригинальных данных


https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isnull.html

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.mean.html


| Код                           | Что делает                                                  |
| ----------------------------- | ----------------------------------------------------------- |
| `df.isnull()`                 | Создаёт таблицу булевых значений (True, если пропуск)       |
| `.mean()`                     | Вычисляет среднее (True=1, False=0), то есть долю пропусков |
| `* 100`                       | Переводит в проценты                                        |
| `missing_ratio.get('col', 0)` | Безопасно достаёт значение даже если столбца нет            |
| `fillna(df['col'].median())`  | Заполняет медианой                                          |


````python
df = pd.read_csv("housing.csv")  # укажи путь к файлу, если не в одной папке
````

````python
# Считаем долю пропусков в процентах
missing_ratio = df.isnull().mean() * 100

# Проверяем конкретный столбец
if missing_ratio.get('total_bedrooms', 0) > 30:
    df = df.drop(columns=['total_bedrooms'])
    print("Столбец 'total_bedrooms' удалён (слишком много пропусков).")
else:
    df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].median())
    print("Пропуски в 'total_bedrooms' заполнены медианой.")
````

Простая стратегия: заполнение медианой для числовых признаков. Это безопасная стратегия для EDA и базового моделирования.
(Подробнее: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html)


````python

# 6) Заполним пропуски медианой для числовых столбцов (копия данных)
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
df_num = df[num_cols].copy()

print('Числовые колонки:', num_cols)
print('Число пропусков до заполнения:')
print(df_num.isna().sum())

df_num_filled = df_num.fillna(df_num.median())

print('\nЧисло пропусков после заполнения (проверка):')
print(df_num_filled.isna().sum())
````

- **Pearson** измеряет линейную зависимость.  
- **Spearman** — ранговая корреляция, устойчива к нелинейностям ( монотонные связи ).  
- **Kendall** — ещё один ранговый коэффициент, полезен для малого количества данных.
(Подробнее: https://en.wikipedia.org/wiki/Correlation_coefficient)


````python

# 10) Корреляционные матрицы
pearson = df_num_filled.corr(method='pearson')
spearman = df_num_filled.corr(method='spearman')
kendall = df_num_filled.corr(method='kendall')

print('Pearson (top 5 correlations with target):')
print(pearson[target].abs().sort_values(ascending=False).head(10))

plt.figure(figsize=(10,8))
sns.heatmap(pearson, cmap='coolwarm', annot=False)
plt.title('Pearson correlation heatmap')
plt.show()
````

## 📉 Линейные зависимости

````python
#📉 Линейные зависимости

target = 'median_house_value'
features = ['total_rooms', 'total_bedrooms', 'households']
for feat in features:
    X = df[[feat]].dropna()
    y = df.loc[X.index, target]
    model = LinearRegression().fit(X, y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=X[feat], y=y, alpha=0.5)
    plt.plot(X, y_pred, color='red')
    plt.title(f'{feat} vs {target} (R²={r2:.3f})')
    plt.show()
````


## 🧩 Нелинейные подгонки (примеры)
Примеры: квадратичная подгонка, лог-трансформация, полиномиальная степень 7 как proxy для сложных зависимостей.


````python

# 11) Топ-6 признаков по абсолютной корреляции с target
top_feats = pearson[target].abs().drop(target).sort_values(ascending=False).head(6).index.tolist()
top_feats
````

````python

# Визуализация: scatter + линейная регрессия для каждого из топ-признаков
for feat in top_feats:
    X = df_num_filled[[feat]].values
    y = df_num_filled[target].values
    model = LinearRegression().fit(X,y)
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=df_num_filled[feat], y=y, alpha=0.3)
    # линия по отсортированным X
    order = np.argsort(X[:,0])
    plt.plot(X[order,0], y_pred[order], color='red')
    plt.title(f'{feat} vs {target} (R²={r2:.3f})')
    plt.xlabel(feat); plt.ylabel(target)
    plt.show()
````

````python

# 12) Нелинейные подгонки — несколько примеров
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

nonlinear_examples = [
    ('population', 'quadratic', 2),
    ('median_income', 'log-transform', 1),
    ('longitude', 'poly7', 7)
]

for feat, kind, deg in nonlinear_examples:
    X_raw = df_num_filled[[feat]].values
    y = df_num_filled[target].values
    if kind == 'log-transform':
        X = np.log1p(np.abs(X_raw))
    else:
        X = X_raw.copy()
    poly = PolynomialFeatures(degree=deg, include_bias=False)
    X_poly = poly.fit_transform(X)
    model = LinearRegression().fit(X_poly, y)
    y_pred = model.predict(X_poly)
    r2 = r2_score(y, y_pred)
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=X_raw.flatten(), y=y, alpha=0.3)
    order = np.argsort(X_raw.flatten())
    plt.plot(X_raw.flatten()[order], y_pred[order], color='red')
    plt.title(f'{feat} — {kind} (deg={deg}) R²={r2:.3f}')
    plt.xlabel(feat); plt.ylabel(target)
    plt.show()
````
